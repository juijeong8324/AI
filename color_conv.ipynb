```python
from PIL import Image, ImageEnhance
```

```python
import glob
```

```python
j = 1
```

```python
images = glob.glob('/home/ubuntu/Yolo_mark/x64/Release/data/img/*.png') # 모든 *.png 파일이 배열로 반환됨
```

```python
### 채도 변화 
```

```python
for i in images:
    print(i)
    test_img = Image.open(i)
    change_color = ImageEnhance.Color(test_img)
    color_output = change_color.enhance(2.5)
    color_output.save('color3_%d.png' %j)
    j = j + 1
```

```python
###밝기 변화 
```

```python
k = 1
```

```python
im2 = glob.glob('/home/ubuntu/Yolo_mark/x64/Release/data/img/*.png') 
```

```python
for i in im2:
    print(i)
    test_img = Image.open(i)
    change_brightness = ImageEnhance.Brightness(test_img)
    br_output = change_brightness.enhance(0.8)
    br_output.save('br1_%d.png' %k)
    k = k + 1
```

```python
for i in images:
    print(i)
    test_img = Image.open(i)
    change_color = ImageEnhance.Color(test_img)
    color_output = change_color.enhance(0.5)
    color_output.save(i)
```

```python
###밝기 변화 2!!이때 k=1, im2 어쩌구 코드 누르고 다음 코드를 실행해주어야 한다. 
```

```python
for i in im2:
    print(i)
    test_img = Image.open(i)
    change_brightness = ImageEnhance.Brightness(test_img)
    br_output = change_brightness.enhance(1.5)
    br_output.save('br2_%d.png' %k)
    k = k + 1
```
